# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Loading the Dataset

```python
df = pd.read_csv('student_scores.csv',  encoding = 'unicode_escape')
```

# Data Cleaning

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
df.head()
```

# Exploratory Data Analysis

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title ('Gender Distribution')    
plt.show()    
    
```

NOTE : FROM THE ABOVE GRAPH WE CAN SEE THAT THE NUMBER OF FEMALES IN THE DATA IS MORE THAN THE NUMBER OF MALES

```python
gb = df.groupby(['ParentEduc']).agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
gb
```

```python
plt.figure(figsize = (4,4))
sns.heatmap (gb, annot =True)
plt.title ("Relationship between Parent's Education and Student's Score")
plt.show()
```

NOTE: FROM THE ABOVE GRAPH WE CAN SEE THAT THE PARENT'S EDUCATION HAVE A GOOD IMPACT ON THEIR SCORES

```python
gb1 = df.groupby(['ParentMaritalStatus']).agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
gb1
```

```python
plt.figure(figsize = (4,4))
sns.heatmap (gb1, annot =True)
plt.title ("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

NOTE: FROM THE ABOVE GRAPH WE CAN SEE THAT THERE IS NO/NEGLIGIBLE IMPACT ON THE STUDENT'S SCORE DUE TO THEIR PARENT'S MARITAL STATUS

```python
groupA = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE = df.loc[(df['EthnicGroup'] == 'group E')].count()

l = ['group A', 'group B', 'group C', 'group D', 'group E']
mlist = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
plt.pie(mlist, labels = l, autopct = '%1.2f%%')
plt.title ('Distibution of Ethnic Groups')
plt.show()
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x ='EthnicGroup')
for bars in ax.containers:
     ax.bar_label(bars)
plt.show()
```

NOTE: FROM THE ABOVE GRAPH WE CAN SEE THAT THE GROUP C HAVE MORE NUMBERS OF ETHNIC COUNT AS COMPARED TO OTHER GROUPS

```python
gb2 = df.groupby(['WklyStudyHours']).agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
gb2
```

```python
plt.figure(figsize = (4,4))
sns.heatmap (gb2, annot =True)
plt.title ("Relationship between Wkly Study Hours and Student's Score")
plt.show()
```

NOTE: FROM THE ABOVE GRAPH WE CAN SEE THAT THE STUDENTS WHO STUDY MORE THAN 10 HOURS WEEKLY GET SCORED GOOD MARKS AS COMPARED TO OTHER STUDENTS

```python
gb3 = df.groupby(['PracticeSport']).agg({'MathScore': 'mean', 'ReadingScore': 'mean', 'WritingScore': 'mean'})
gb3
```

```python
plt.figure(figsize = (4,4))
sns.heatmap (gb3, annot =True)
plt.title ("Relationship between PracticeSport and Student's Score")
plt.show()
```

NOTE: FROM THE ABOVE GRAPH WE CAN SEE THAT THERE IS A FEW IMAPCT ON STUDENT'S SCORE DUE TO  PRACTICING THE SPORT
